## Univariate analysis

```python
plt.Figure(figsize=(8,5))
plt.hist(x='mpg',data=car_design)
plt.xlabel('mpg')
plt.ylabel('frequency')
plt.title('MPG Distribution',size=15)
plt.show()
```

```python
sns.countplot(x='am',y=None,data=car_design)
plt.title('Automatic cars Vs Manual cars')
plt.figure(figsize=(5,4))
plt.show
```

```python
car_design['cyl'].value_counts()
```

```python
import plotly.express as px
```

```python
plt.figure(figsize=(9,7))
plt.pie(x=car_design['cyl'].value_counts(),labels=[8,4,6],explode=[0.02,0.02,0.02],shadow=True,autopct="%.2f%%")
plt.show()
```

```python
plt.boxplot(x='mpg',data=car_design)
plt.title('mpg Distribution')
plt.show()
```

###  2. Bivariate analysis

```python
car_design.head()
```

```python
sns.scatterplot(x='wt',
    y='mpg',data=car_design)
plt.title('WT Vs MPG')
plt.show()
```

```python
sns.barplot(x='cyl',
    y='mpg',data=car_design)
plt.title('No. of Cyl Vs Mpg')
plt.show()
```
